# Calculate Fourier Transform

```python
import numpy as np
import matplotlib.pyplot as plt
```

Eval function is used to evaluate an expression that results in a text string or a numeric value. The Eval function evaluates the string expression and returns its value.

```python
# Get input sequence from the user as a space-separated string
input_sequence_str = input("Enter the input sequence (space-separated values): ")

# Split the input string by spaces and convert it to a list of numbers
xn = [float(item) for item in input_sequence_str.split()]

# Now, 'xn' contains the user-entered sequence as a list of numbers
print("Input sequence:", xn)
```

fft.fft function represents

FFT (Fast Fourier Transform) refers to a way the discrete Fourier Transform (DFT) can be calculated afficiently, by using symmation in the

calculated terms

It converts a signal into individual spectral components and thereby provides frequency information about the signal

```python
xk=np.fft.fft(xn)
xk
```

```python
PSD= np.abs(xk)
PSD
```

```python
plt.stem(PSD)
plt.xlabel('k---->')
plt.ylabel('x(k)---->')
plt.title('PSD')
```

PSD of Sine Signal

```python
F1= 20
F2= 40
fs=500
t= np.arange(0,2, (1/fs))
plt.figure(figsize=(15,3))
x1= np.sin(2*np.pi*F1*t)
plt.plot(t,x1)
```

```python
plt.figure(figsize= (15,3))
x2= np.sin(2*np.pi*F2*t)
plt.plot(t,x2)
```

```python
plt.figure(figsize= (15,3))
xn=x1+ 2*x2
plt.plot(t,xn)
```

```python
xk= np.fft.fft(xn)
PSD= np.abs(xk)
plt.plot(PSD)
```

```python
N=len(xn)
K= np.arange(N)*(fs/N)
print(N)
#print(K)
```

matplotlib.pyplot stem() creates stem plots. A Stem plot plots vertical lines at each x position covered under the grapn from the baseline tox and places a marker there.

```python
plt.figure(figsize=(15,3))
plt.stem(K[0:int(N/2)], PSD[0: int(N/2)])
plt.xlabel('k---->')
plt.ylabel('X(k)---->')
plt.title('PSD')
```
