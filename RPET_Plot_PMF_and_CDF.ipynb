# Plot the Probability Mass Function (PMF) and Cumulative Distribution Function (CDF) of the number of heads when tossing two coins

```python
import scipy.special
import matplotlib.pyplot as plt

x = int(input("Number of coins you want the PDF for: "))
l = []
l1 = []
cx = 0
CX = []

for i in range(x + 1):
    l.append(scipy.special.comb(x, i) / 2**x)
    cx = cx + (scipy.special.comb(x, i) / 2**x)
    CX.append(cx)
    l1.append(i)

print("Distribution Table")
print("x | p(x) | c(x)")
for i in range(len(l1)):
    print(f"{i} | {l[i]} | {CX[i]}")

# Plotting the PMF
plt.bar(l1, l, color="blue")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.title("PMF of Heads for tossing two coins")
plt.show()

# Plotting the CDF
plt.bar(l1, CX, color="green")
plt.xlabel("Number of Heads")
plt.ylabel("Cumulative Probability")
plt.title("CDF of Heads for tossing two coins")
plt.show()
```

# Calculating and Plotting the Probability Mass Function (PMF) of the number of heads when tossing two coins

```python
import scipy.special
import matplotlib.pyplot as plt

x = int(input('Number of coins you want the PMF for: '))

# Initialize lists to store PMF values and cumulative probabilities
l1 = []  # Number of heads
l = []   # PMF values
cx = 0   # Cumulative probability
CX = []  # Cumulative probabilities

# Calculate the PMF and cumulative probabilities for each possible outcome
for i in range(x + 1):
    pmf = scipy.special.comb(x, i) / 2**x  # Calculate PMF
    l.append(pmf)

    cx += pmf  # Update cumulative probability
    CX.append(cx)

    l1.append(i)

print("Distribution Table")
print("x | p(x) | c(x)")

for i in range(len(l1)):
    print(f"{l1[i]} | {l[i]} | {CX[i]}")

# Plotting the PMF
plt.bar(l1, l)
plt.xticks(l1, range(x + 1))
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.title("PMF of Heads for Tossing 2 Coins")
plt.show()

# Plotting the CDF
plt.bar(l1, CX)
plt.xticks(l1, range(x + 1))
plt.xlabel("Number of Heads")
plt.ylabel("Cumulative Probability")
plt.title("CDF of Heads for Tossing 2 Coins")
plt.show()
```
