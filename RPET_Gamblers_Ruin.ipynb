# Simulate a gambling scenario where you start with a certain amount of money (gambling_money) and aim to reach a specific goal (gambling_goal) by making random bets.

```python
import numpy as np
```

```python
import random

def gamblers_ruin(starting_money, goal):
    gambling_money = starting_money
    gambling_sim = []  # List to track the gambling simulation

    while gambling_money > 0 and gambling_money < goal:
        # Make a random bet, -1 for loss, 1 for win
        w_or_l = random.choice([-1, 1])
        gambling_money += w_or_l
        gambling_sim.append(gambling_money)

    return gambling_sim

starting_money = 100
goal = 200
simulation_results = gamblers_ruin(starting_money, goal)
print("Simulation Results:", simulation_results)
```

```python
import matplotlib.pyplot as plt
plt.figure (figsize=(15,6))
plt.plot(gamblers_ruin(50,100))
plt.yticks(np.arange(-20,120,10))
plt.axhline(y=0, color='r')
plt.axhline(y=100, color='b')
plt.xlabel("Number of bets")
plt.ylabel("Money")
plt.title("Gambling Simulations")
```
