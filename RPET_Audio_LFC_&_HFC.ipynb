# Observe frequency component of audio 8 file using PSD and afterthe low and high frequency components. Observethe frequency component of audio 8 files using Power spectral density

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
fs= 1000 #sample frequency

# Generate time vector properly
t= np.arange(1000)/fs
signal_a= np.sin(2*np.pi*100*t) # with frequency of 100
plt.figure(figsize=(20,4))
plt.plot(t,signal_a)
plt.xlabel("Time--->")
plt.ylabel("Amplitude--->")
plt.title("Sine signal with frequency 100 Hz")
```

```python
f2= abs(np.fft.fft(signal_a))
print(f2.shape)
print()
plt.figure(figsize=(15,4))
plt.stem(f2[0:500])
plt.ylabel("Magnitude--->")
plt.xlabel("Frequency--->")
plt.title("PSD of Sine signal")
```

```python
signal_b= np.sin(2*np.pi*20*t) # with frequency of 20
plt.figure(figsize=(20,4))
plt.plot(t,signal_b)
plt.xlabel("Time--->")
plt.ylabel("Amplitude--->")
plt.title("Sine signal with frequency 20 Hz")
```

```python
plt.figure(figsize=(15,4))
f1= abs(np.fft.fft(signal_b))
plt.stem(f1[0:100])
plt.xlabel("Frequency--->")
plt.ylabel("Magnitude--->")
plt.title("PSD of Sine signal")
```

```python
plt.figure(figsize=(15,4))
signal_c = signal_a + signal_b

#print(signal_c)
#print(t.shape)
#print()
#print(t)

plt.plot(t[0:300], signal_c[0:300])
plt.xlabel("Time--->")
plt.ylabel("Amplitude--->")
plt.title("Addition of 2 Sine signal")
```

```python
plt.figure(figsize=(15,4))

plt.plot(t[0:500], signal_c[0:500])
plt.xlabel("Time--->")
plt.ylabel("Amplitude--->")
plt.title("Addition of 2 Sine signal")
```

```python
plt.figure(figsize=(15,4))
f3= abs(np.fft.fft(signal_c))
plt.stem(f3[0:300])
plt.xlabel("Frequency--->")
plt.ylabel("Magnitude--->")
plt.title("PSD of Addition of 2 Sine signal")
```

# Filter the low and high frequency of components using Low Pass Filter and High Pass Filter(System)


```python
fc= 30 #Cut off frequencyof the filter
w1=fc/(fs/2)  #Normalize the frequency
b,a= signal.butter(5,w1,'low')  #System
output= signal.filtfilt(b,a,signal_c)
plt.figure(figsize=(20,4))
plt.plot(t,output)
plt.xlabel("Time--->")
plt.ylabel("Amplitude--->")
plt.title("Sine signal with low frequency component filtered")
```

```python
f4= abs(np.fft.fft(data))
plt.figure(figsize=(15,4))
plt.plot(f4[0:int(len(f4)/2)])
plt.xlabel("Frequency--->")
plt.ylabel("Magnitude--->")
plt.title("PSD of StarWars3.wav audio file")
```

```python
fc=1000
w=fc/(fs/2)
b,a= signal.butter(5,w,'low')
output= signal.filtfilt(b,a,data)
```

```python
plt.figure(figsize=(15,4))
plt.plot(output)
```

```python
w,h= signal.freqz(b,a)
plt.plot(w,abs(h))
plt.xlabel("Frequency--->")
plt.ylabel("Amplitude--->")
plt.title("Magnitude Response")
```
