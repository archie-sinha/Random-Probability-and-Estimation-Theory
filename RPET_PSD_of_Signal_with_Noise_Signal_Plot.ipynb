# Write a program to calculate psd of a random signal. Add a noise signal with original signal and plot it.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
```

```python
fs= 1000 #sample frequency

# Generate time vector properly
t= np.arange(1000)/fs
signal_a= np.sin(2*np.pi*150*t) # with frequency of 150
plt.figure(figsize=(25,3))
plt.plot(t,signal_a)
plt.xlabel("Time: ")
plt.ylabel("Amplitude: ")
plt.title("Sine signal with frequency 150 Hz")
```

```python
f2= abs(np.fft.fft(signal_a))
print(f2.shape)
print()
plt.figure(figsize=(15,4))
plt.stem(f2[0:300])
plt.ylabel("Magnitude: ")
plt.xlabel("Frequency: ")
plt.title("PSD of Sine signal")
```
