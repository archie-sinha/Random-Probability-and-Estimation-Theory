# ARCHISHA SINHA

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a range of x values from 0 to 10*pi with 1024 data points
x = np.linspace(0, 10*np.pi, 1024)

# Calculate the corresponding y values by taking the sine of x
y = np.sin(x)

# Create a plot of x versus y
plt.plot(x, y)

# Label the x-axis
plt.xlabel("time --->")

# Label the y-axis
plt.ylabel("magnitude --->")

# Set the title of the plot
plt.title("Sine Signal")

# Display the plot
plt.show()
```
